```python
import re

def extract_abstract(text: str) -> str:
    pattern = r'abstract(.*?)(?=keywords:)'
    match = re.search(pattern, text, re.IGNORECASE | re.DOTALL)

    if match:
        return match.group(1).strip()
    return ""
```

```python
print(extract_abstract(proposal_text))
```
